```python
rates = dict()
```

```python
for o in p.offers:
    key = f"{o.have}-{o.want}"
    if key not in rates:
        rates[key] = list()
    rates[key].append(o.conversion_rate)
```

```python
for key, value in rates.items():
    if "Chaos Orb" in key:
        print("\n\n" + key)
        df = pd.DataFrame(value)
        df.columns = [key]
        print(df.describe().transpose())
        df.hist(bins=5)
```
